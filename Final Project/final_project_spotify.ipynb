```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

plt.rcParams["figure.figsize"] = (20,10)
```

```python
spotify_df = pd.read_csv('Data\spotify\charts.csv')
# print(spotify_df)
```

```python
# extract the duration, and position for songs from 2021

spotify_df = spotify_df[(spotify_df['date'].str.contains('2021'))]
spotify_df = spotify_df[['duration', 'position', 'country', 'explicit']]
#convert duration to seconds
spotify_df['duration'] = spotify_df['duration'].divide(1000)
spotify_df = spotify_df[spotify_df['country'] == 'us']
spotify_df = spotify_df[spotify_df['position'] <= 30]

# print(spotify_df)
print(len(spotify_df))
plt.scatter(spotify_df['duration'], spotify_df['position'], c=spotify_df['explicit'])
#make a legend for the explicit column
plt.legend(['Explicit', 'Not Explicit'], loc='upper right')
plt.xlabel('Duration (seconds)')
plt.ylabel('Position')
plt.title('Duration vs Position')
plt.show()
```
